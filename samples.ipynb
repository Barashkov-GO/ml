# Plot doubled

```python
import random
import pandas as pd
from utilities import date_range

a = [random.randint(100, 120) for _ in range(100)]
b = ["A" if random.randint(0, 100) >= 50 else "B" for _ in range(100)]

c = [random.randint(10, 20) for _ in range(100)]
d = [random.randint(30, 40) for _ in range(100)]
e = [random.randint(70, 80) for _ in range(100)]

c1 = [random.randint(10, 20) for _ in range(100)]
d1 = [random.randint(30, 40) for _ in range(100)]
e1 = [random.randint(70, 80) for _ in range(100)]

it = ['Sell' for _ in range(50)] + ['Buy' for _ in range(50)]
dates = date_range('2024-01-01', '2024-10-01', freq=1)
dates_first = [i for i, j in dates][:100]


df = pd.DataFrame(
    {"a": a, "b": b, "c": c, "d": d, "e": e, "c1": c1, "d1": d1, "e1": e1, 'it': it, 'date': dates_first}
)
df
```

## Bars
